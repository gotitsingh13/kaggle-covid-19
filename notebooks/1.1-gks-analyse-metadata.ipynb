## metadata analysis

```python
# author analysis

author_publication_distribution = dict()


for i, row in df.iterrows():
    if isinstance(row["authors"], str):
        authors = row['authors'].split(";")

        for pos, author in enumerate(authors):
            author = author.strip()
            if author_publication_distribution.get(author):
                if author_publication_distribution[author].get(pos):
                    author_publication_distribution[author][pos] += 1
                else:
                    author_publication_distribution[author][pos] = 1
            else:
                author_publication_distribution[author] = {pos: 1}
```

```python
author_publication_distribution
```

```python
write_json_file(file_path=DATA_INTERIM + "author_publication_distribution.json", 
                data=author_publication_distribution)

# author_publication_distribution = read_json_file(file_path=DATA_INTERIM + "author_publication_distribution.json")
```

```python
## Top authors by publications
authors_publications = {x: y for x, y in sorted(author_publication_distribution.items(), 
                                                key=lambda item: sum(item[1].values()),
                                               reverse=True)}
```
